```python
import sklearn.datasets
import pandas as pd
import numpy as np
import umap
```

```python
pendigits = sklearn.datasets.load_digits()
mnist = sklearn.datasets.fetch_openml('mnist_784')
fmnist = sklearn.datasets.fetch_openml('Fashion-MNIST')
```

```python
type(mnist)
```

mnist (jeu de données MNIST avec des chiffres manuscrits à 28 x 28 pixels) téléchargé depuis [openML](https://www.openml.org/)

MNIST est un jeu de données de référence pour la classification de chiffres manuscrits. Chaque image est une matrice de 28x28 pixels, représentant des chiffres de 0 à 9.
L’ensemble contient 70 000 images (60 000 pour l’entraînement et 10 000 pour le test), ce qui le rend idéal pour évaluer la précision et la robustesse des modèles de reconnaissance d’images.

Fashion-MNIST est un ensemble de 70 000 images en niveaux de gris, également de 28x28 pixels, représentant des articles de mode (t-shirts, pantalons, chaussures, etc.)

```python
import umap.umap_ as umap
```

```python
mapper = umap.UMAP().fit(pendigits.data)
```

# Larger Datasets

```python
mapper = umap.UMAP().fit(fmnist.data)
```

```python
umap.plot.points(mapper)
```

```python
umap.plot.points(mapper, labels=fmnist.target, theme='fire')
```

```python
mapper = umap.UMAP().fit(fmnist.data[:30000])
```

```python
hover_data = pd.DataFrame({'index':np.arange(30000),
                           'label':fmnist.target[:30000]})
hover_data['item'] = hover_data.label.map(
    {
        '0':'T-shirt/top',
        '1':'Trouser',
        '2':'Pullover',
        '3':'Dress',
        '4':'Coat',
        '5':'Sandal',
        '6':'Shirt',
        '7':'Sneaker',
        '8':'Bag',
        '9':'Ankle Boot',
    }
)
```

```python
umap.plot.output_notebook()
```

```python
p = umap.plot.interactive(mapper, labels=fmnist.target[:30000], hover_data=hover_data, point_size=2)
umap.plot.show(p)
```

```python
umap.plot.connectivity(mapper, show_points=True)
```
