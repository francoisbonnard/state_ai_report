# Import Necessary Packages

```python
import torch
import torchvision.transforms as T
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
from transformers import DetrForObjectDetection, DetrImageProcessor
```

# Load the Model

# Get Image from Internet

```python
https://raw.githubusercontent.com/bpbpublications/Building-Transformer-Models-with-PyTorch/blob/main/chapter%207/tulip_field.png
```

```python
url = "https://raw.githubusercontent.com/bpbpublications/Building-Transformer-Models-with-PyTorch/main/chapter7/tulip_field.png"
response = requests.get(url)
img = Image.open(BytesIO(response.content))
img = img.convert("RGB")
```

```python
# Process the image
```

```python
transform = T.Compose([
    T.Resize(800),
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])
img_tensor = transform(img).unsqueeze(0)
```

# Perform the Prediction

```python
with torch.no_grad():
    outputs = model(img_tensor)
target_sizes = torch.tensor([img.size[::-1]])
results = processor.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=0.9)[0]


```

# Plot the Prediction

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.imshow(img)

# is a qualitative colormap containing 20 distinct colors.
# These colors are then used in the visualization code to assign
# unique colors to different object categories detected in the image.
colors = plt.get_cmap("tab20").colors

# results["scores"] is prediction
# results["labels"] is true label
# results["boxes"] is the bounding box of object
for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    x, y, w, h = box
    w = w - x
    h = h - y
    rect = plt.Rectangle((x, y), w, h, linewidth=1, edgecolor=colors[label % 20], facecolor="none")
    ax.add_patch(rect)
    ax.text(x, y, f"{model.config.id2label[label.item()]} {round(score.item(), 3)}", fontsize=15, color=colors[label % 20])

plt.show()
```
